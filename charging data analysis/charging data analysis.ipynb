```python
df['Order creation time'] = df['Order creation time'].str.strip() #Remove line breaks from the string
```

```python
df['Order creation time'] = pd.to_datetime(df['Order creation time'],errors='coerce')
```

```python
invalid_dates = df['Order creation time'].isna().sum()  #check if there is any invalid data
print(invalid_dates)
```

```python
duplicate_count = df.duplicated().sum()         #Duplicate value check
```

```python
duplicate_count
```

```python
#time feature extration
df['year'] = df['Order creation time'].dt.year
df['month'] = df['Order creation time'].dt.month
df['day'] = df['Order creation time'].dt.day
df['hour'] = df['Order creation time'].dt.hour
df['weekday'] = df['Order creation time'].dt.day_name()
```

```python
df['End Time'] = pd.to_datetime(df['End Time'])
df['Start Time'] = pd.to_datetime(df['Start Time'])
df['Payment time'] = pd.to_datetime(df['Payment time'])
```

```python
df['Charging Duration'] = (df['End Time'] - df['Start Time']).dt.total_seconds() / 3600 #charging duratioin(hour)
```

```python
with pd.option_context('display.max_columns', None):
    display(df.head())                                     #display all columns
```

```python
# outliers treatment
df = df[df['Transaction power/kwh'] >= 0]
df = df[df['Actual Payment/Yuan'] >= 0]
df = df[(df['Temperature(℃)'] >= -25) & (df['Temperature(℃)'] <= 55)]
df = df[(df['Relative Humidity(%)'] >=25) & (df['Relative Humidity(%)'] <= 95)]
```

```python
df['year'].unique()
```

```python
#Data visualization
```

```python
# charging power trend(per month)
df['year_month'] = df['Order creation time'].dt.to_period('M').astype(str)
month_power = df.groupby('year_month')['Transaction power/kwh'].sum().reset_index()
```

```python
plt.figure(figsize = (12,6))
sns.lineplot(data = month_power,x='year_month',y='Transaction power/kwh',marker='o',color='green')
plt.title('Charging Power Trend(per month)')
plt.xlabel('Year_month')
plt.ylabel('Power(Kwh)')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
# Total charging power of each district
plt.figure(figsize=(6,6))
district_power = df.groupby('District Name')['Transaction power/kwh'].sum().sort_values(ascending=False).reset_index()
sns.barplot(data=district_power,
            x='District Name',
            y='Transaction power/kwh',
            hue='District Name',
            palette='coolwarm',
            legend=False,
            dodge=False)
plt.title('Total charging power of each district')
plt.xlabel('District Name')
plt.ylabel('Total Power(Kwh)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Distribution of end causes
plt.figure(figsize=(12,6))
end_causes = df['end_cause'].value_counts().reset_index()
end_causes.columns = ['end_cause','counts']
sns.barplot(data=end_causes,
            x='counts',
            y='end_cause',
           hue='counts',
           palette='Set2',
           legend=False,
           dodge=False)
plt.title('Distribution of end causes')
plt.xlabel('counts')
plt.ylabel('end_cause')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#average charging power per day(Box diagram)
df['day'] = df['Order creation time'].dt.date
daily_power = df.groupby('day')['Transaction power/kwh'].mean().reset_index()

plt.figure(figsize=(12,6))
sns.boxplot(data=daily_power,
            y='Transaction power/kwh',
           color='skyblue')
plt.title('average charging power per day')
plt.ylabel('average charging power(Kwh)')
plt.tight_layout()
plt.show()
```

```python
#Hourly Power Distribution by Charging Location
df['Power/kw'] = df['Transaction power/kwh']/df['Charging Duration']
df['Hour'] = df['Order creation time'].dt.hour
df.replace({r'\s+': ' '}, regex=True, inplace=True)
df_sum = df.groupby(['Hour', 'Location Information']).agg(
                                                        mean_power=('Power/kw', 'mean'),
                                                        sum_power=('Power/kw', 'sum')
                                                          )
df_sum = df_sum.reset_index()

pivot_table = df_sum.pivot_table(
    columns='Hour',
    index='Location Information',
    values='mean_power',
)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=False, fmt=".5f", cmap="Blues", linewidths=0.5, cbar_kws={'label': 'Power (kW)'})
plt.title('Hourly Power Distribution by Charging Location')
plt.xlabel('Hour')
plt.ylabel('Charging Station')
plt.tight_layout()
plt.show()
```

```python
#Charging difference between weekdays and weekends
df['weekend'] = df['weekday'].isin(['Saturday','Sunday'])
df['order_count'] = 1
weekday_summary = df.groupby('weekend').agg({
    'order_count': 'count',
    'Transaction power/kwh': ['sum', 'mean']
}).reset_index()
weekday_summary.columns = ['weekend', 'order_count', 'total_power_kwh', 'avg_power_kwh']
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

sns.barplot(data=weekday_summary, x='weekend', y='order_count', ax=axs[0], hue='weekend', palette='pastel', legend=False)
axs[0].set_title('Comparison of Order Quantity')
axs[0].set_ylabel('Order Quantity')
axs[0].set_xticks([0, 1])
axs[0].set_xticklabels(['Weekday', 'Weekend'])

sns.barplot(data=weekday_summary, x='weekend', y='total_power_kwh', ax=axs[1], hue='weekend', palette='pastel', legend=False)
axs[1].set_title('Total Charging Power')
axs[1].set_ylabel('Total Power (kWh)')
axs[1].set_xticks([0, 1])
axs[1].set_xticklabels(['Weekday', 'Weekend'])

sns.barplot(data=weekday_summary, x='weekend', y='avg_power_kwh', ax=axs[2], hue='weekend', palette='pastel', legend=False)
axs[2].set_title('Average Charging Power')
axs[2].set_ylabel('Average Power (kWh)')
axs[2].set_xticks([0, 1])
axs[2].set_xticklabels(['Weekday', 'Weekend'])

plt.suptitle('Charging Behavior: Weekday vs Weekend', fontsize=16)
plt.tight_layout()
plt.show()
```
